```python
#Part 1: b<a+1, stable spiral
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

a=0.5
b=0.5
N=20000
x=np.linspace(0.01,N/100,N)
#print(x)

y1=np.zeros(len(x))
y2=np.zeros(len(x))

for i in range(len(x)):
  y1[i]=(b+1)/(a*x[i])-1/(a*x[i]*x[i])
  y2[i]=b/(a*x[i])
plt.title('Plot of X nullclines and Y nullclines')  
plt.plot(x,y1,label='X nullcline')
plt.plot(x,y2,label='Y nullcline')
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.ylim((-0.1,8))
plt.show()
```

```python
A=0.8
B=1.2
fA=5
fB=5
fC=5
while(True):
  C=(A+B)/2
  fA=(b+1)/(a*A)-1/(a*A*A)-b/(a*A)
  fB=(b+1)/(a*B)-1/(a*B*B)-b/(a*B)
  fC=(b+1)/(a*C)-1/(a*C*C)-b/(a*C)
  C=(A+B)/2
  if(abs(fC)<0.005):
    break
print(C)
x_fixed=C
y_fixed1=b/(a*C)
y_fixed2=(b+1)/(a*C)-1/(a*C*C)
print(y_fixed1)
print(y_fixed2)
```

```python
xx=[0]*N
yy=[0]*N
t=np.linspace(0.01,N/100,N)
dt=0.01
xx[0]=1
yy[0]=5.01
for i in range(N-1):
  xx[i+1]=xx[i]+(1-(b+1)*xx[i]+a*xx[i]*xx[i]*yy[i])*dt
  yy[i+1]=yy[i]+(b*xx[i]-a*xx[i]*xx[i]*yy[i])*dt
plt.plot(xx,yy,'red')
plt.plot(x,y1,'blue')
plt.plot(x,y2,'green')
plt.title('Plot for a stable spiral')
#plt.xlim((-0.,0.005))
plt.xlim((-1,6))
plt.ylim((-1,7.5))
plt.show()
plt.plot(t,xx,'red',label='x(t)')
#plt.show()
plt.title('Position vs Time Graph')
plt.plot(t,yy,'blue',label='y(t)')
plt.xlabel('Time')
plt.ylabel('Position')
plt.legend()
plt.show()
```

```python
flag=0
for i in range(25,N-1):
    if(yy[i]>yy[i-1] and yy[i]>yy[i+1] and flag==0):
        t1=t[i]
        flag=flag+1
    elif(yy[i]>yy[i-1] and yy[i]>yy[i+1] and flag==1):
        t2=t[i]
        break
time_period=t2-t1
print(time_period)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#w = 3
Y, X = np.mgrid[-1:7.5:100j, -1:6:100j]
U=(1-(b+1)*X+a*X*X*Y)
V=(b*X-a*X*X*Y)
#speed=np.sqrt(U*U+V*V)

fig = plt.figure(figsize=(12, 12))
gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[2, 2, 2])
ax0 = fig.add_subplot(gs[0, 0])
ax0.streamplot(X, Y, U, V,color='aqua')
ax0.set_title('Flow for Brusselator for b<a+1')
plt.plot(x,y1,'green',label='X nullcline')
plt.plot(x,y2,'blue',label='Y nullcline')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.plot(xx,yy,'red',label='Flow towards the stable fixed point.')
plt.legend()
plt.ylim(0,7.5)
plt.xlim(0,6)
plt.show()
```
